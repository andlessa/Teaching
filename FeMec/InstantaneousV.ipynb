```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML
rc('animation', html='jshtml')
import seaborn as sns

#Define plotting style:
sns.set() #Set style
sns.set_style('ticks',{'font.family':'serif', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.5)
```

```python
tmin = 0.
tmax = 10.

xF = lambda t: t**2 - 0.1*t**3
fig = plt.figure(figsize=(10,7))
tpts = np.linspace(tmin,tmax,100)
line1 = plt.plot(tpts,xF(tpts),linewidth=4)
line2, = plt.plot(tpts[1::30],xF(tpts[1::30]),'o--',linewidth=2,markersize=15)
xM, = plt.plot(tpts[1::30],[0.]*len(tpts[1::30]),'|-',linewidth=2,markersize=15)
plt.xlabel('t (s)')
plt.ylabel('x (m)')
```

```python
def init():
    line2.set_data([], [])
    xM.set_data([], [])
    return (line2,xM)
```

```python
def animate(i):
    xpts = np.linspace(tmin,tmax,i)
    ypts = xF(xpts)
    line2.set_data(xpts, ypts)
    xM.set_data(xpts,[0.]*len(xpts))
    return (line2,xM)
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=50, interval=1000, 
                               blit=True)
anim
```
